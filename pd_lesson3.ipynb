```python
np.random.seed(100)
my_data = np.random.randint(0, 101, (4, 3))
my_data

my_row_index = ['CA', 'NY', 'AZ', 'TX']  # 4 rows
my_col_index = ['Jan', 'Feb', 'Mar']  # 3 columns
full_index = pd.DataFrame(data=my_data, index=my_row_index, columns=my_col_index)
full_index = pd.DataFrame(data=my_data, index= ['CA', 'NY', 'AZ', 'TX'], columns=['Jan', 'Feb', 'Mar'])

full_index[['Jan']]  # df
full_index['Jan']  # series
full_index[['Jan', 'Feb']]
full_index.loc['CA']  # series
full_index.loc[['CA']]  # df
full_index.iloc[:, 0]  # series
full_index.iloc[:, 0].to_frame()  # df
full_index.loc[full_index['Jan'] > 80, ['Feb']]
full_index.loc[full_index['Jan'] > 80, 'Jan':'Feb']  # more common!

full_index.iloc[1 : 2, 1 : 2]
full_index.iloc[[True, True, False, True]]
full_index.iloc[list((full_index['Jan'] > 80).values), 0 : 1]  # less common! 
```
