```python
tips = pd.read_csv(r"tips.csv")
df_tips = tips.copy()

def measure_tip(total_bill, tip):
    if tip >= total_bill * 0.25:
        return "Generous"
    else:
        return "Not Generous"

# level 1
df_tips['tipper'] = np.vectorize(measure_tip)(df_tips['total_bill'], df_tips['tip'])
# df_tips = df_tips.sort_values(by='tipper')

# level 2 -- bonus*
def calc_on_row(row):
    #print(type(row))
    #print(row)
    if row['tip'] >= row['total_bill'] * 0.25:
        return "Generous"
    else:
        return "Not Generous"    
#display(df_tips.head())
df_tips['tipper'] = df_tips.apply(calc_on_row, axis=1)

# level 3 -- bonus*
df_tips['tipper'] = df_tips.apply(lambda row: "Generous" if row['tip'] >= row['total_bill'] * 0.25 else "Not Generous", axis=1)
df_tips = df_tips.sort_values(by='tipper')
df_tips.head(11)
```
